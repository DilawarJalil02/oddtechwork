```python
import pandas as pd
import json
```

```python
df = pd.read_csv('dataset_6.csv')
```

```python
df.head(10)
```

```python
df['data.rawTranscriptions'] = df['data.rawTranscriptions'].apply(lambda x: json.loads(x))
```

```python
df.head()
```

```python
df['data.id'] = df['data.id'].astype(int)
```

```python
df.head()
```

```python
data_timestamp = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        data_timestamp.append(df['data.timestamp'][i])
```

```python
data_number = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        data_number.append(df['data.number'][i])
```

```python
data_application = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        data_application.append(df['data.application'][i])
```

```python
data_id = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        data_id.append(df['data.id'][i])
```

```python
audio_name = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        audio_name.append(df['data.application'][i] + '-' + df['data.id'][i].astype(str) + '-in.wav')
```

```python
chunk_name = []

for i in range(len(df['data.rawTranscriptions'])):
    for j in range(len(df['data.rawTranscriptions'][i])):
        chunk_name.append(df['data.application'][i] + '-' + df['data.id'][i].astype(str) + '-in-' + str(j+1) + '.wav')
```

```python
data_raw_transcriptions = []

for k in range(len(df['data.rawTranscriptions'])):
    for l in range(len(df['data.rawTranscriptions'][k])):
        data_raw_transcriptions.append(df['data.rawTranscriptions'][k][l])
```

```python
print(len(data_timestamp),
      len(data_number),
      len(data_application),
      len(data_id),
      len(audio_name),
      len(chunk_name),
      len(data_raw_transcriptions))
```
